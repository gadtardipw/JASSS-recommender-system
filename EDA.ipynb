# Exploratory Data Analysis and Additional Data Cleaning for JASSS Dataset

```python
# Importing the dataset, cleaning the pre-known issue that some keywords are delineated with semi-colons (;) rather than commas (,)

import pandas as pd
import chardet

# detect the encoding 
with open('jasss_dataset.csv', 'rb') as f:
    encoding_result = chardet.detect(f.read())
    
# read the CSV file with the detected encoding
df = pd.read_csv('jasss_dataset.csv', sep='|', encoding=encoding_result['encoding'])


# replace semicolons with commas in the keywords column
df['Keywords'] = df['Keywords'].str.replace(';', ',')
```

```python

df.head(10)
```

```python
# overview for columns and data types
df.info()

# analyze missing data
missing_data = df.isnull().sum()
print("Missing data:\n", missing_data)

# calculate the percentage of missing data in each column
missing_percentage = (missing_data / len(df)) * 100
print("\nMissing data percentage:\n", missing_percentage)

df.head(10)
```

```python
# Additonal cleaning

# remove rows with null values in 'Title', 'Abstract', and 'Keywords'
df = df.dropna(subset=['Title', 'Abstract', 'Keywords'])

# remove rows with abstracts less than 20 characters long
df = df[df['Abstract'].str.len() >= 20]


# reset the index
df.reset_index(drop=True, inplace=True)

# clean the keywords by removing whitespace, removing single-letter keywords, and filtering out empty strings
df['Keywords'] = df['Keywords'].apply(lambda x: ', '.join([keyword.strip() for keyword in x.split(',') if len(keyword.strip()) > 1]))

# remove rows where the keyword is 'KEYWORDS'
df = df[df['Keywords'].apply(lambda x: x != 'KEYWORDS' if isinstance(x, str) else True)]

# drop duplicate rows based on 'Title' and 'Abstract'
df = df.drop_duplicates(subset=['Title', 'Abstract'])


df.info()

# analyze missing data
missing_data = df.isnull().sum()
print("Missing data:\n", missing_data)

# calculate the percentage of missing data in each column
missing_percentage = (missing_data / len(df)) * 100
print("\nMissing data percentage:\n", missing_percentage)

df.head(10)
```

```python
# calculate the average length of title, abstract, number of authors, and number of keywords
avg_title_length = df['Title'].dropna().str.split().apply(lambda x: len(x)).mean()
avg_abstract_length = df['Abstract'].dropna().str.split().apply(lambda x: len(x)).mean()
avg_num_authors = df['Authors'].dropna().str.split(',').apply(len).mean()
avg_num_keywords = df['Keywords'].dropna().str.split(',').apply(len).mean()

# display the averages in a table
averages_table = pd.DataFrame({
    'Average Title Length (words)': [avg_title_length],
    'Average Abstract Length (words)': [avg_abstract_length],
    'Average Number of Authors': [avg_num_authors],
    'Average Number of Keywords': [avg_num_keywords]
})
averages_table
```

```python
# Publications per year

# convert the 'Publish_date' column to datetime format
df['Publish_date'] = pd.to_datetime(df['Publish_date'])

# group by year and count the number of publications
yearly_publications = df.groupby(df['Publish_date'].dt.year).size()

# visualize the distribution of publication dates
yearly_publications.plot(kind='bar', title='Number of Publications per Year')
```

```python
# Number of unique authors and keywords

# find the number of unique names
unique_authors = df['Authors'].str.split(', ', expand=True).stack().nunique()
print(f'Number of unique authors: {unique_authors}')

# find the number of unique keywords
unique_keywords = df['Keywords'].dropna().str.split(', ', expand=True).stack().nunique()
print(f'Number of unique keywords: {unique_keywords}')
```

```python
# Top authors

# extract all authors and count their occurrences
authors = df['Authors'].str.split(', ').explode().str.strip()
top_authors = authors.value_counts()

# display the top authors and their publication count
top_authors.head(20).plot(kind='bar', title='Top 20 Authors by Publication Count')
```

```python
# Average length of titles, keywords, abstracts by year

import numpy as np
import matplotlib.pyplot as plt


# convert publish_date to datetime and extract the year
df['Year'] = pd.to_datetime(df['Publish_date']).dt.year

# calculate the number of words in titles, abstracts, and the number of keywords
df['Title_Words'] = df['Title'].apply(lambda x: len(x.split()))
df['Abstract_Words'] = df['Abstract'].apply(lambda x: len(x.split()))
df['Num_Keywords'] = df['Keywords'].apply(lambda x: len(x.split(', ')))

# group by year and calculate the mean lengths
mean_lengths_by_year = df.groupby('Year').agg({
    'Title_Words': np.mean,
    'Abstract_Words': np.mean,
    'Num_Keywords': np.mean
}).reset_index()

# plot the average number of words in titles per year
plt.figure(figsize=(15, 5))
plt.bar(mean_lengths_by_year['Year'], mean_lengths_by_year['Title_Words'])
plt.xlabel('Year')
plt.ylabel('Average Title Words')
plt.title('Average Number of Words in Titles by Year')
plt.xticks(mean_lengths_by_year['Year'], rotation=45)
plt.show()

# plot the average number of words in abstracts per year
plt.figure(figsize=(15, 5))
plt.bar(mean_lengths_by_year['Year'], mean_lengths_by_year['Abstract_Words'])
plt.xlabel('Year')
plt.ylabel('Average Abstract Words')
plt.title('Average Number of Words in Abstracts by Year')
plt.xticks(mean_lengths_by_year['Year'], rotation=45)
plt.show()

# plot the average number of keywords per year
plt.figure(figsize=(15, 5))
plt.bar(mean_lengths_by_year['Year'], mean_lengths_by_year['Num_Keywords'])
plt.xlabel('Year')
plt.ylabel('Average Number of Keywords')
plt.title('Average Number of Keywords by Year')
plt.xticks(mean_lengths_by_year['Year'], rotation=45)
plt.show()
```

```python
# Top keywords

import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter 

# extract keywords
keywords_list = df['Keywords'].apply(lambda x: x.split(', '))

# excluded words
exclude_words = {
    "Agent-Based Modeling",
    "Agent-Based Model",
    "Agent-Based Modelling",
    "Agent-Based Simulation",
    "Simulation",
    "Agent Based Modeling",
    "Agent-Based Models",
    'ABM',
    'Agents',
    'Agent-Based Social Simulation',
    'Agent-Based',
    'Agent Based Modelling',
    'Modeling',
    'Modelling'
}

# count keyword occurrences excluding the specified terms
keyword_counts = Counter()
for keywords in keywords_list:
    filtered_keywords = [kw for kw in keywords if kw not in exclude_words]
    keyword_counts.update(filtered_keywords)

# get the top n keywords
top_n_keywords = 50
top_keywords = keyword_counts.most_common(top_n_keywords)

# plot the top keywords
plt.figure(figsize=(15, 8))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=[count for word, count in top_keywords], y=[word for word, count in top_keywords])
ax.set_xlabel("Frequency")
ax.set_ylabel("Keywords")
ax.set_title(f"Top {top_n_keywords} Keywords Excluding Generic Terms (e.g. Agent Based Modeling, ABM, Simulation) ")
plt.show()
```

```python
# Wordcloud of keywords

from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis')
wordcloud.generate_from_frequencies(keyword_counts)

# display the word cloud
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# co-authorship network to see how tightly-knit community is

import networkx as nx
import matplotlib.pyplot as plt
import itertools

# create a co-authorship network
def create_coauthorship_network(df):
    G = nx.Graph()
    
    for authors in df['Authors'].str.split(', '):
        for author1, author2 in itertools.combinations(authors, 2):
            if G.has_edge(author1, author2):
                G[author1][author2]['weight'] += 1
            else:
                G.add_edge(author1, author2, weight=1)
    
    return G
G = create_coauthorship_network(df)

# draw the co-authorship network with labels (change font_size to change name label size)
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, node_size=20, edge_color="grey", width=0.5)
nx.draw_networkx_labels(G, pos, font_size=5, font_family="sans-serif")
plt.title("Co-authorship Network")
plt.show()
```
